# **Course Registration System**(8)

```python
# Course class
class Course:
    def __init__(self, course_id, name, credits, max_seats):
        self.course_id = course_id
        self.name = name
        self.credits = credits
        self.max_seats = max_seats
        self.enrolled_students = []

    def has_seat(self):
        return len(self.enrolled_students) < self.max_seats

    def course_summary(self):
        return f"{self.course_id} - {self.name} | Credits: {self.credits} | Enrolled: {len(self.enrolled_students)}/{self.max_seats}"


# Student class
class Student:
    def __init__(self, student_id, name, max_credits):
        self.student_id = student_id
        self.name = name
        self.max_credits = max_credits
        self.courses = []

    def total_credits(self):
        return sum(course.credits for course in self.courses)

    def student_summary(self):
        course_list = [course.name for course in self.courses]
        return f"Student: {self.name} | Credits: {self.total_credits()}/{self.max_credits} | Courses: {course_list}"


# Registration System
class RegistrationSystem:
    def enroll(self, student, course):
        # Prevent duplicate registration
        if course in student.courses:
            print("Already enrolled in this course.")
            return

        # Check credit limit
        if student.total_credits() + course.credits > student.max_credits:
            print("Credit limit exceeded.")
            return

        # Check seat availability
        if not course.has_seat():
            print("No seats available.")
            return

        # Enroll student
        student.courses.append(course)
        course.enrolled_students.append(student)
        print(f"{student.name} enrolled in {course.name}")


# ----- Example Usage -----
course1 = Course("CS101", "Intro to CS", 3, 2)
course2 = Course("MA101", "Math", 4, 1)

student1 = Student("S1", "Alice", 6)
student2 = Student("S2", "Bob", 6)

system = RegistrationSystem()

system.enroll(student1, course1)
system.enroll(student2, course1)
system.enroll(student2, course2)
system.enroll(student1, course2)  # No seats available

# Reports
print("\n--- Student Summary ---")
print(student1.student_summary())
print(student2.student_summary())

print("\n--- Course Summary ---")
print(course1.course_summary())
print(course2.course_summary())
```
